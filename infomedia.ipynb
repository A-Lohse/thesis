```python
#dates = [datetime.strptime(d, "%Y-%m-%d") for d in list(dates)]
dateend = datetime(2020,9,1)
dstart = datetime(2020,8,20)
linde_date = datetime.strptime("2020-08-26", "%Y-%m-%d")
```

```python
plt.figure(figsize=(12, 8)) 

sns.lineplot(y = "metoo",x = "dato", label = "Mentions of #Metoo",data= df ,alpha = 0.8, marker= ".", markersize=14, linewidth=3.5)
plt.ylabel("Daily mentions of #Metoo in the media")
#plt.title("Mentions of #Metoo in the media")
plt.xlabel("Date")
plt.xticks(rotation = 90)

plt.axvline(x = linde_date, color = 'r', label = 'Sofie Linde\'s speech', linestyle = "--",linewidth=2.5) 
plt.xlim(linde_date - timedelta(30),linde_date + timedelta(30))
#plt.ylim(0,1.2)
plt.legend()

plt.savefig("plots\\infomedia_30.png",bbox_inches = "tight")

plt.show()
```
